```python
from diffusers import StableDiffusionPipeline
```

```python
import torch
```

```python
connection=StableDiffusionPipeline.from_pretrained("rupeshs/LCM-runwayml-stable-diffusion-v1-5")
```

```python
connection=connection.to("cuda")
```

```python
prompt="a man with a car"
```

```python
image = connection(prompt).images[0]
```

```python
image.save('output.png')
```
